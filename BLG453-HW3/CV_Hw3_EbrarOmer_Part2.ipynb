```python
import numpy as np
import cv2
from scipy import io
import scipy
import imageio
import os
import glob
from sklearn.metrics import precision_score
```

```python
my_images_list=[]
ground_truth_list=[]
for image in glob.glob('results/*'):
    input_img = cv2.imread(image)
    my_images_list.append(input_img)
```

```python
for image in glob.glob('CATS-master/output/bsds/test/sing_scale_test//*'):
    input_img = cv2.imread(image)
    ground_truth_list.append(input_img)
```

```python
precision_score_list=[]
for i in range (len(ground_truth_list)):
    my_img=my_images_list[i]
    db_img=ground_truth_list[i]
    score = (precision_score(my_img.flatten(), db_img.flatten(), average='micro', zero_division=1))
    print(score)
    precision_score_list.append(score)
```

```python
print(sum(precision_score_list)/len(precision_score_list))
```
