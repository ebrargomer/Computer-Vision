```python
import numpy as np
import cv2
from scipy import io
import scipy
import imageio
import os
import glob
from sklearn.metrics import precision_score
```

```python
read_path  = os.path.join('BSR_full/BSR/BSDS500/data/groundTruth/test')
write_path = os.path.join('results')
```

```python
my_images_list=[]
ground_truth_list=[]
for image in glob.glob('BSR_full/BSR/BSDS500/data/images/test/*'):
    input_img = cv2.imread(image)
    img_name = image.split("test\\",1)[1]
    path=("myresults/")
    gray = cv2.cvtColor(input_img, cv2.COLOR_BGR2GRAY)
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)
    img1 = cv2.Canny(blurred, 100,200)
    img2 = cv2.Canny(blurred, 100,300)
    img3 = cv2.Canny(blurred, 200, 300)
    img4 = cv2.Canny(blurred, 200, 400)
    img5 = cv2.Canny(blurred, 300, 400)
    img6 = cv2.Canny(blurred, 300, 500)
    final_img=cv2.addWeighted(img1, 0.5, img2, 0.5,0)
    cv2.addWeighted(final_img,0.5,img3,0.5,0)
    cv2.addWeighted(final_img,0.5, img4,0.5,0) 
    cv2.addWeighted(final_img,0.5,img5,0.5,0)
    cv2.addWeighted(final_img,0.5, img6,0.5,0)     
    #In order to obtain a higher precision score, 
    #I applied canny to the image at different intervals and collected it with weighted.
    cv2.imwrite(path+img_name,final_img)
    my_images_list.append(final_img)
```

```python
for filename in os.listdir(read_path):
        data = io.loadmat(os.path.join(read_path,filename))
        edge_data_255=np.empty(data['groundTruth'][0][0][0][0][0].shape)
        for i in range (data['groundTruth'].shape[1]):
            edge_data = data['groundTruth'][0][i][0][0][1]*255
            edge_data_255+=edge_data
        new_img_name = filename.split('.')[0]+'.jpg'
        ground_truth_list.append(edge_data_255)
        imageio.imwrite(os.path.join(write_path,new_img_name), edge_data_255)
```

```python
precision_score_list=[]
for i in range (len(ground_truth_list)):
    my_img=my_images_list[i]
    db_img=ground_truth_list[i]
    score = (precision_score(my_img.flatten(), db_img.flatten(), average='micro', zero_division=1))
    print(score)
    precision_score_list.append(score)
```

```python
print(sum(precision_score_list)/len(precision_score_list))
```
